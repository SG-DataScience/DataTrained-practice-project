World Happiness Report:

The World Happiness Report is a landmark survey of the state of global happiness. The first report was published in 2012, the second in 2013, the third in 2015, and the fourth in the 2016 Update. The World Happiness 2017, which ranks 155 countries by their happiness levels, was released at the United Nations at an event celebrating International Day of Happiness on March 20th. The report continues to gain global recognition as governments, organizations and civil society increasingly use happiness indicators to inform their policy-making decisions. Leading experts across fields – economics, psychology, survey analysis, national statistics, health, public policy and more – describe how measurements of well-being can be used effectively to assess the progress of nations. The reports review the state of happiness in the world today and show how the new science of happiness explains personal and national variations in happiness.

```python
#finding out the best random state
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import r2_score
```

```python
max_score=0
maxrs=0
for i in range(1,200):
    x_train, x_test, y_train, y_test = train_test_split(x2, y, test_size=0.30, random_state=i)
    lm=LinearRegression()
    lm.fit(x_train,y_train)
    pred=lm.predict(x_test)
    score=r2_score(y_test,pred)
        
    if score>max_score:
        max_score=score
        maxrs=i
        
print('best score is ',max_score,'on random state',maxrs)    
```

```python
# therefore at random state 124 the score is highest,so using this random statr we will make the train test split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x2, y, test_size=0.30, random_state=124)
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
print('coefficient:',lm.coef_)
print('intercept:',lm.intercept_)
print('score:',lm.score(x_train,y_train))
pred=lm.predict(x_test)
print('\n')
print('Predicted happiness score:',pred)
print('actual happiness score:',y_test)
```

```python
print('error:')

print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean squared error:',mean_squared_error(y_test,pred))

print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
from sklearn.metrics import r2_score
print('for linear regression r2 score is:',r2_score(y_test,pred))
```

```python
#using other regression algorithms
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)
pred=svr.predict(x_test)
print('error:')

print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean squared error:',mean_squared_error(y_test,pred))

print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,pred)))
print('\n')
print('r2 score:',r2_score(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
pred=dtr.predict(x_test)
print('error:')

print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean squared error:',mean_squared_error(y_test,pred))

print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,pred)))
print('\n')
print('r2 score:',r2_score(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
pred=rfr.predict(x_test)
print('error:')

print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean squared error:',mean_squared_error(y_test,pred))

print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,pred)))
print('\n')
print('r2 score:',r2_score(y_test,pred))
```

```python
from sklearn.model_selection import cross_val_score
scr=cross_val_score(lm,x2,y,scoring='r2',cv=5)
print('Cross validation score of LinearRegression model:',scr.mean())
```

```python
from sklearn.model_selection import cross_val_score
scr=cross_val_score(svr,x2,y,scoring='r2',cv=5)
print('Cross validation score of SVR:',scr.mean())
```

```python
from sklearn.model_selection import cross_val_score
scr=cross_val_score(dtr,x2,y,scoring='r2',cv=5)
print('Cross validation score of DecisionTreeRegressor:',scr.mean())
```

```python
from sklearn.model_selection import cross_val_score
scr=cross_val_score(rfr,x2,y,scoring='r2',cv=5)
print('Cross validation score of RandomForestRegressor:',scr.mean())
```
